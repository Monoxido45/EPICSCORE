# Printing and obtaining all results for quantile regression and regression:

```python
import os
import pandas as pd

original_path = os.getcwd()

def custom_table_generator(result):
    metrics = list(result.columns)[1:]

    print(f"\\begin{{tabular}}{{l{'c' * len(metrics)}}}\\toprule")
    print("Dataset & " + " & ".join(metrics) + "\\\\\\midrule")
    for _, row in result.iterrows():
        values = [row["Dataset"]]
        for metric in result.columns[1:]:
            value = row[metric]
            if pd.isna(value):
                values.append("-")
            else:
                values.append(f"{value}")
        print(" & ".join(values) + "\\\\")
    print("\\bottomrule\\end{tabular}")
```

Getting the results:

```python
folder_path = os.path.dirname(original_path) + "/Experiments_code/results"
```
